## Step 2: Variables & Data Types

This is the absolute bedrock of Python. If you can't store data, you can't analyze it.

### Definition

A Variable is a reserved memory location to store values. In other words, a variable in a python program gives data to the computer for processing.

### Description (The Analyst's View)

Think of variables as labeled boxes in a warehouse.

- The Variable Name is the label on the outside of the box (e.g., total_sales).
- The Value is what you put inside the box (e.g., 5000).
- The Data Type is the kind of object the box can hold (numbers, text, or true/false logic).

Unlike Excel, where you see the data in a grid cell (like A1), in python, you must give that data a name so you can call it later.

### Examples

Here are 5 examples illustrating the most common data types you will encounter in analysis: Integers, Floats, Strings, Booleans, and Dynamic Typing.

#### Example 1: Integers (int) Used for whole numbers

In data analysis, this is used for counts (e.g., number of customers, number of items sold).

```python
# We create a box labeled 'items_sold' and put the number 150 inside.
items_sold = 150

# We can perform math on it immediately
projected_sales = items_sold * 2

print(items_sold)
print(projected_sales)
```


#### Example 2: Floats (float) Used for numbers with decimals

This is crucial for financial data, percentages, and scientific measurements.

```python
# 'float' is short for floating-point number
transaction_amount = 99.50
tax_rate = 0.08

# Calculate total with tax
total_cost = transaction_amount + (transaction_amount * tax_rate)

print(total_cost)
```

#### Example 3: Strings (str) Used for text

This includes customer names, product categories, ID codes, or addresses. You must always wrap text in quotes ("" or '').

```python
customer_name = "Alice Smith"
product_category = 'Electronics'

# You can combine (concatenate) strings using the + sign
label = customer_name + " bought " + product_category

print(label)
```

#### Example 4: Booleans (bool) Used for logic: True or False

This is essential for filtering data later (e.g., "Is this customer active?").

```python
is_premium_member = True
has_churned = False

# We can check the type of the variable using type()
print(is_premium_member)
print(type(is_premium_member))
```

#### Example 5: Variable Re-assignment (Dynamic Typing)

In Python, the boxes are flexible. You can take a number out and put text in (though you should be careful doing this!).

```python
# Initially, the variable holds a number
current_status = 100
print(current_status)

# Now we update it to hold text
current_status = "Processing Complete"
print(current_status)

# This is useful when cleaning data (e.g., replacing a messy number with a category)
```
