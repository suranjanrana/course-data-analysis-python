## 2. Variables & Data Types

This is the absolute bedrock of Python. If you can't store data, you can't analyze it.

### Definition

A Variable is a reserved memory location to store values. In other words, a variable in a python program gives data to the computer for processing.

### Description (The Analyst's View)

Think of variables as labeled boxes in a warehouse.

- The Variable Name is the label on the outside of the box (e.g., total_sales).
- The Value is what you put inside the box (e.g., 5000).
- The Data Type is the kind of object the box can hold (numbers, text, or true/false logic).

Unlike Excel, where you see the data in a grid cell (like A1), in python, you must give that data a name so you can call it later.

### Examples

Here are 5 examples illustrating the most common data types you will encounter in analysis: Integers, Floats, Strings, Booleans, and Dynamic Typing.

#### Example 1: Integers (int) Used for whole numbers

In data analysis, this is used for counts (e.g., number of customers, number of items sold).

```python
# We create a box labeled 'items_sold' and put the number 150 inside.
items_sold = 150

# We can perform math on it immediately
projected_sales = items_sold * 2

print(items_sold)
print(projected_sales)
```


#### Example 2: Floats (float) Used for numbers with decimals

This is crucial for financial data, percentages, and scientific measurements.

```python
# 'float' is short for floating-point number
transaction_amount = 99.50
tax_rate = 0.08

# Calculate total with tax
total_cost = transaction_amount + (transaction_amount * tax_rate)

print(total_cost)
```

#### Example 3: Strings (str) Used for text

This includes customer names, product categories, ID codes, or addresses. You must always wrap text in quotes ("" or '').

```python
customer_name = "Alice Smith"
product_category = 'Electronics'

# You can combine (concatenate) strings using the + sign
label = customer_name + " bought " + product_category

print(label)
```

#### Example 4: Booleans (bool) Used for logic: True or False

This is essential for filtering data later (e.g., "Is this customer active?").

```python
is_premium_member = True
has_churned = False

# We can check the type of the variable using type()
print(is_premium_member)
print(type(is_premium_member))
```

#### Example 5: Variable Re-assignment (Dynamic Typing)

In Python, the boxes are flexible. You can take a number out and put text in (though you should be careful doing this!).

```python
# Initially, the variable holds a number
current_status = 100
print(current_status)

# Now we update it to hold text
current_status = "Processing Complete"
print(current_status)

# This is useful when cleaning data (e.g., replacing a messy number with a category)
```

## 3. Lists
In Data Analysis, you rarely work with one number at a time. You work with thousands. We need structures to hold these collections.


### Definition

A List is an ordered, mutable (changeable) collection of items. It is defined by square brackets `[]`.

### Description

Think of a List as a column in Excel.

- It keeps data in a specific order.
- You can access items by their position (called the Index).

**Important**: Python uses "0-based indexing." This means the first item is at position 0, the second is at position 1, etc.

### Examples

#### Example 1: Creating a List

Storing a sequence of monthly sales figures.

```python
# A list of integers
sales_2023 = [150, 200, 250, 300, 320]

# A list of strings (Categories)
departments = ["Electronics", "Clothing", "Home", "Garden"]

print(sales_2023)
```

#### Example 2: Indexing (Accessing Data)

Retrieving specific items using their position.

```python
# Remember: Count starts at 0!
first_month = sales_2023[0]  # Gets 150
third_month = sales_2023[2]  # Gets 250

# Negative indexing lets you count from the end (-1 is the last item)
last_month = sales_2023[-1] # Gets 320

print(first_month, last_month)
```

#### Example 3: Slicing (Getting a Subset)

Extracting a chunk of data (e.g., Q1 sales). The syntax is `[start : stop]`. Note that the 'stop' number is *excluded*.

```python
# Get items from index 0 up to (but not including) 3
q1_sales = sales_2023[0:3] 

print(q1_sales) 
# Output: [150, 200, 250]
```

#### Example 4: Modifying the List

Lists are "mutable," meaning we can change them after creation.

```python
# We realized a mistake in the data. The first month was actually 160.
sales_2023[0] = 160

# We need to add a new month (Append adds to the end)
sales_2023.append(350)

print(sales_2023)
```

#### Example 5: Simple Analysis Functions

Python has built-in math functions that work directly on lists

```python
prices = [10.50, 20.00, 5.00, 100.00]

min_price = min(prices)
max_price = max(prices)
total_revenue = sum(prices)
number_of_txns = len(prices) # 'len' calculates length

print(total_revenue)
```

## 4. Dictionaries

 ### Definition
 
 A Dictionary is an unordered collection of items stored in `Key: Value` pairs. It is defined by curly braces `{}`.

### Description

Think of a Dictionary as a single row of data or a specific Customer Record

- Instead of using a numbered index (0, 1, 2) to find data, you use a meaningful Key (like "Name", "Age", "City").
- This is exactly how JSON data works, which is a common format for web data.

### Examples

#### Example 1: Creating a Dictionary

Defining a single transaction record.

```python
transaction = {
    "txn_id": "TXN-001",
    "amount": 450.00,
    "customer": "John Doe",
    "is_valid": True
}

print(transaction)
```

#### Example 2: Accessing Data (Lookups)

You pull data out by asking for the Key.

```python
# What was the amount?
sale_value = transaction["amount"]

# Who was the customer?
buyer = transaction["customer"]

print(buyer, "spent", sale_value)
```

#### Example 3: Adding and Updating Data

Like lists, dictionaries can be changed.

```python
# Update an existing value
transaction["amount"] = 500.00 

# Add a BRAND NEW key-value pair
transaction["discount_applied"] = 0.10

print(transaction)
```

#### Example 4: Inspecting Keys and Values

Sometimes you need to see a list of all columns (keys) or all data points (values).

```python
# specific methods .keys() and .values()
columns = transaction.keys()
data = transaction.values()

print(list(columns))
```

#### Example 5: The "Dataset" Structure (List of Dictionaries)

This is the most common data structure you will see before we start using Pandas. It represents a table: A list (rows) containing dictionaries (columns)

```python
sales_data = [
    {"date": "2023-01-01", "sales": 100},
    {"date": "2023-01-02", "sales": 150},
    {"date": "2023-01-03", "sales": 200}
]

# Accessing the sales of the second day
# First get list item 1 (the second row), then get key "sales"
day_2_sales = sales_data[1]["sales"]

print(day_2_sales)
```

## Logic & Loops

### Definition

- **Logic (Control Flow):**
    - Code that makes decisions. It executes specific blocks of code only if certain conditions are met (True/False).
- **Loops (Iteration):**
    - Code that repeats a specific block of actions for every item in a sequence (like a list).

### Description (The Analyst's View)

- **Logic (if/else):**
    - Think of this as the "Filter" or "Calculated Column" in Excel. You use it to categorize data (e.g., "If sales > 1000, mark as 'High Priority', otherwise 'Standard'").
- **Loops (for):**
    - Think of this as a robot that goes through your rows of data one by one. Instead of manually copying and pasting 100 times, you tell the loop to "do this action for every row."

### Examples

#### Example 1: Conditional Logic (if, elif, else)

Making a decision based on the value of a variable.

```python
sales = 1500

if sales > 2000:
    category = "High Performance"
elif sales > 1000:
    category = "Medium Performance"
else:
    category = "Low Performance"

print(f"This sale is: {category}")
# Output: This sale is: Medium Performance
```

#### Example 2: The for Loop

Iterating through a list. This is the most common loop you will use.

```python
# A list of daily revenues
daily_revenue = [100, 200, 300, 400]

# "For every 'amount' inside the list 'daily_revenue'..."
for amount in daily_revenue:
    # ...print the amount plus a label
    print("Revenue:", amount)
```

#### Example 3: Logic inside Loops (Filtering)

This combines the two concepts. We loop through all data, but only "keep" or process the ones that meet our criteria.

```python
temperatures = [22, 30, 18, 35, 20]
hot_days = [] # Empty list to store results

for temp in temperatures:
    if temp > 25:
        # If it's hot, add it to our list
        hot_days.append(temp)

print(f"Hot days were: {hot_days}")
# Output: Hot days were: [30, 35]
```

#### Example 4: The Accumulator Pattern

Calculating a running total. This helps you understand how functions like sum() actually work under the hood.

```python
sales = [10, 20, 30, 40]
total_sales = 0 # Start at zero

for sale in sales:
    total_sales = total_sales + sale
    # On loop 1: 0 + 10 = 10
    # On loop 2: 10 + 20 = 30, etc.

print(f"Final Total: {total_sales}")
```

#### Example 5: List Comprehensions (The "Pro" Shortcut

Python allows you to write a loop in a single line. This is very popular in data analysis for cleaning data quickly.

```python
raw_prices = [10, 20, 30, 40]

# Syntax: [ operation  for  item  in  list ]
# We want to add tax (10%) to every price
taxed_prices = [price * 1.10 for price in raw_prices]

print(taxed_prices)
# Output: [11.0, 22.0, 33.0, 44.0]
```
