## Boolean Masking

````python
arr = np.array([10, 15, 20, 30, 35, 40, 41, 43, 50])

print(arr)
````

````python
mask = arr > 30
````

````python
arr[mask]
````

````python
arr[arr % 5 != 0]
````

````python
# Variation for Data Analysis: Filtering rows based on column values
sales_data = np.array([
    [101, 50, 200],  # ID, Units Sold, Price
    [102, 10, 50],
    [103, 75, 300],
    [104, 20, 100],
    [105, 60, 250]
])
````

````python
sales_data[[2, 4], 0]
````

````python
sales_data[:, 1]
````

````python
total_sales = sales_data[:, 2] * sales_data[:, 1]
print(total_sales)
````

````python
# Price less than or equals to 100
sales_data[sales_data[:, 2] <= 100]
````

````python
total = total_sales.sum()
print(total)
````
