````python
df_excel["time_of_day"].drop_duplicates()
````

````python
morning_df = df_csv[df_excel['time_of_day'] == 'Morning']
afternoon_df = df_csv[df_excel['time_of_day'] == 'Afternoon']
night_df = df_csv[df_excel['time_of_day'] == 'Night']

morning_df_2 = morning_df.groupby('date').agg(morning_sales=('money', 'sum'))
afternoon_df_2 = afternoon_df.groupby('date').agg(afternoon_sales=('money', 'sum'))
night_df_2 = night_df.groupby('date').agg(night_sales=('money', 'sum'))


time_df = pd.merge(morning_df_2, afternoon_df_2, on='date', how='outer')
pd.merge(time_df, night_df_2, on='date', how='outer')
````

````python
morning_df_2.join(afternoon_df_2, how="outer").join(night_df_2, how="outer")
````

````python

Morning= df_excel[df_excel["time_of_day"] == "Morning"].groupby("date").agg(morning_sales=("money", "sum"))
Afternoon= df_excel[df_excel["time_of_day"] == "Afternoon"].groupby("date").agg(afternoon_sales=("money", "sum"))
Night= df_excel[df_excel["time_of_day"] == "Night"].groupby("date").agg(night_sales=("money", "sum"))

Morning.join(Afternoon, how = "outer").join(Night, how = "outer")
````

### pd.concat()

````python
sales_2024 = (
    df_excel[(df_excel["year"] == 2024) & (df_excel["month_name"].isin(["Mar", "Apr"]))]
        .groupby(["year", "month_name"])
        .agg(total_sales=("money", "sum"))
        .reset_index()
)

sales_2024_2 = (
    df_excel[(df_excel["year"] == 2024) & (df_excel["month_name"].isin(["May", "Jun"]))]
        .groupby(["year", "month_name"])
        .agg(total_sales=("money", "sum"))
        .reset_index()
)

sales_2025 = (
    df_excel[(df_excel["year"] == 2025) & (df_excel["month_name"].isin(["Jan", "Feb", "Mar"]))]
        .groupby(["year", "month_name"])
        .agg(total_sales=("money", "sum"))
        .reset_index()
)

sales_2024
````

````python
concat_df = pd.concat([sales_2024, sales_2025, sales_2024_2])
concat_df
````
